```python
from pathlib import Path
import random, matplotlib.pyplot as plt
from PIL import Image

root = Path("../data/processed/images")   # <-- always valid now
jpgs = list(root.rglob("*.jpg"))
print(f"Found {len(jpgs):,} JPGs")

k = min(6, len(jpgs))
sample = random.sample(jpgs, k)
fig, axes = plt.subplots(2, 3, figsize=(8, 6))
for ax, img_path in zip(axes.ravel(), sample):
    ax.imshow(Image.open(img_path), cmap="gray")
    ax.set_title(img_path.stem.split("_")[0])   # Gardiner code
    ax.axis("off")
plt.tight_layout()
```
